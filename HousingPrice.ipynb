```python
sns.boxplot(x='LotArea', data=hpdf_new)
```

```python
sns.scatterplot(x='LotArea',y='SalePrice',data=hpdf_new)
```

```python
sns.scatterplot(x='LotArea',y='SalePrice',hue='SaleCondition',data=hpdf_new)
```

```python
#To find the strong relation features with SalePrice target variable
hpdf_corr = hpdf_new.corr()
Corr_output = abs(hpdf_corr['SalePrice'])
Corr_output[Corr_output > 0.5]
hpdf_new = hpdf_new[['OverallQual','YearBuilt','YearRemodAdd','TotalBsmtSF','1stFlrSF','GrLivArea','FullBath','TotRmsAbvGrd','GarageYrBlt','GarageCars','GarageArea','SalePrice']]
hpdf_new = hpdf_new.corr()
plt.figure(figsize=(15,10))
sns.heatmap(hpdf_new, annot = True)
```
